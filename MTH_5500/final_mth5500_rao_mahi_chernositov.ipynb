## Monte-Carlo Bond Pricing

Consider the following Vasicek model for the interest rate $(R_t, t\geq0)$ under the risk-neutral probability

$$dR_t = (0.01-0.05R_t)dt + 0.1d\tilde{B_t} \ \ \ \ \ \ R_0 =0.1$$

1. Plot 100 paths of the above process on $[0, 1]$ with a 0.01 discretization.
2. Use the risk-neutral pricing formula to find the price $B(0, 1)$ at time $0$ of a bond with maturity $T = 1$ and $B(1, 1) =100$ by averaging over $1000$ paths. 

```python
# for calculating the itterative expression
def Vasicek(brownian:np.array, time_index: np.array, initial:int, size:int, a:float, b:float, s:float):
    alpha = a
    beta = b
    sigma = s
    
    temp = initial
    new_arr = [initial]*(size+1)
    for i in range(size):
        new_arr[i+1] = temp + ((alpha - beta * temp) * time_index[i]) + (sigma * brownian[i])
        temp += ((alpha - beta * temp) * time_index[i]) + (sigma * brownian[i])
    return np.array(new_arr) 
```

```python
a_plot = np.apply_along_axis(Vasicek, 1, np.delete(brownian_increments, 99, 1), time_increment, 0.1, 99, 0.01, 0.05, 0.1)
```

```python
plt.figure(figsize=(15,5))
plt.title("$dR_t = (0.01-0.05R_t) \ dt + 0.1{B_t}$")
for i in a_plot[:100]:
    plt.plot(i)
```

We have that the price of the zero-coupon bond is:
$$B(t,T)=\tilde{E}[e^{\int_t^T {R_s ds}}| \ F_s]$$

```python
def integrate(arr: np.array, time: np.array):
    return np.trapz(arr, time)
```

```python
def zeroB(rates:np.array, t1:float , t2:float):
    size = a_plot.shape[1]
    price = 0
    
    lower = int(t1 * size)
    top = int(t2 * size)
    price = np.average(np.exp(-np.apply_along_axis(integrate, 1, a_plot[:,lower:top], time_index[lower:top])))
        
    return round(price, 4)
```

```python
B_0_1 = zeroB(a_plot, 0, 1)
```

```python
print("The price of the bond B(0, 1) at time 0 is ${}".format(B_0_1 * 100))
```
